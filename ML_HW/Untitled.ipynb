```python
import os
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import KFold
from sklearn.metrics import classification_report



from joblib import dump
from preprocess import prep_data
from joblib import load

import matplotlib
```

```python
df = pd.read_csv("fish_participant.csv")
df = df.assign(hw = df["Height"] * df["Width"])
df = df.assign(vol = (df["Height"]* df["Width"])*df["Length1"])

```

```python
fish_holdout_2 = pd.read_csv("fish_holdout_demo.csv")
fish_holdout_2 = fish_holdout_2.assign(hw = fish_holdout_2["Height"] * fish_holdout_2["Width"])
fish_holdout_2 = fish_holdout_2.assign(vol = (fish_holdout_2["Height"]* fish_holdout_2["Width"])*fish_holdout_2["Length1"])
fish_holdout_2.to_csv("fish_holdout_2.csv")
```

```python
X, y = prep_data(df)
```

```python
kf = KFold(n_splits = 20, shuffle = True, random_state = 42)
kf.get_n_splits(X)

print(kf)

for train_index, test_index in kf.split(X):
    print("TRAIN:", train_index, "TEST:", test_index)
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
```

```python
wrf = RandomForestClassifier(class_weight = "balanced")
wrf.fit(X_train.astype(int), y_train.astype(int))
y_pred = wrf.predict(X_test)
```

```python
dump(wrf, "wrf_jl.joblib")
```
